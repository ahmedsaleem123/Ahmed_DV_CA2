**i) Identify the differences between a graph and a tree.**<br><br>
**Ans)** Following are given some major differences between a graph and a tree.<br><br>

 - A graph is a data structure that consists of a set of nodes and a set of edges that connect pairs of nodes, where as, a tree is a special type of graph that is undirected, connected, and has no cycles.
<br><br>
 - Edges in a graph can be directed or undirected, where as, all edges in a tree are undirected.
<br><br>
 - A node in a graph can have multiple incoming and outgoing edges, where as,a node in a tree can have only one parent node, and zero or more child nodes.
<br><br>
 - A graph can have multiple connected components, where as,a tree is always connected.
